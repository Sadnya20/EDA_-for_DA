# Data Preprocessing

### Handling duplicate values

```python
df_olympics.duplicated()
```

```python
df_olympics.duplicated().sum()
```

```python
# Duplicates --> Drop
# Missing --> Imputation (Mean or Median for Numerical and Mode for categorical)
# Outlier --> Replacement (Mean or Median for Numerical and Mode for categorical)
```

```python
df_olympics = df_olympics.drop_duplicates()
```

```python
df_olympics.duplicated().sum()
```

### Handling missing values

```python
df_olympics.isna()
```

```python
df_olympics.isnull().sum()
```

```python
round(df_olympics.isnull().sum() / len(df_olympics) * 100, 1)
```

Age --> Median

Height --> Mean

Weight --> Mean

Medal --> No Medal

Region --> Unknown

```python
df_olympics.skew(numeric_only =  True)
```

```python
# If skewness is out of control then use median as replacement strategy else use mean
```

```python
df_olympics['Medal'].mode()
```

```python
df_olympics = df_olympics.fillna(value = {'Age' : df_olympics['Age'].quantile(0.5), #.median()
                                          'Height': df_olympics['Height'].mean(),
                                          'Weight': df_olympics['Weight'].mean(),
                                          'Medal': 'No Medal',
                                          'Region': 'Unknown'})
```

```python
df_olympics.isnull().sum()
```

### Handling outlier values

```python
import numpy as np
```

```python
numeric_data = df_olympics.select_dtypes(include = np.number)
numeric_data
```

```python
for col in numeric_data.drop(columns = 'Year').columns:
    Q1 = df_olympics[col].quantile(0.25)
    Q2 = df_olympics[col].quantile(0.5)
    Q3 = df_olympics[col].quantile(0.75)
    print(f"Processing {col}")
    IQR = Q3 - Q1
    LL = Q1 - (1.5 * IQR)
    UL = Q3 + (1.5 * IQR)
    # Outlier Detection
    outliers = ((df_olympics[col] < LL) | (df_olympics[col] > UL))
    num_outliers = outliers.sum()
    if num_outliers > 0:
        print(f"{col} has {round(num_outliers / len(df_olympics) * 100, 2)}% outliers")
        # Outlier Treatment
        # display(df_olympics[outliers])
        # df_olympics.loc[(df_olympics[col] < LL) | (df_olympics[col] > UL), col] = Q2
    else:
        print(f"{col} has no outliers")
```
